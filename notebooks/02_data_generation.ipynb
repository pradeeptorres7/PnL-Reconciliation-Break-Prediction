```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)

rows = 5000
```

```python
df = pd.DataFrame({

    'DV01': np.random.randint(
        50000,
        250000,
        rows
    ),

    'Vega': np.random.randint(
        10000,
        150000,
        rows
    ),

    'Gamma': np.random.randint(
        1000,
        30000,
        rows
    ),

    'MarketMove': np.random.uniform(
        0.1,
        1.5,
        rows
    ),

    'VaR': np.random.randint(
        100000,
        600000,
        rows
    ),

    'ResidualPnL': np.random.randint(
        1000,
        100000,
        rows
    ),

    'TradeCount': np.random.randint(
        100,
        1000,
        rows
    )

})
df
```

```python
#Business Logic
df['ReconBreak'] = np.where(

(
(df['ResidualPnL'] > 50000)
&
(df['MarketMove'] > 0.8)
)

|
(
df['TradeCount'] > 700
),

1,

0
)
```
